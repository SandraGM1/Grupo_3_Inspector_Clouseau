```python
import sqlite3
from faker import Faker
import random
import pandas as pd
```

```python
cnx = sqlite3.connect("base_datos_proveedores.db")
```

```python
cursor_prov = cnx.cursor()
```

```python
cursor_prov.execute("DROP TABLE IF EXISTS pieza")
```

```python
#Mi lista de constantes
categorias_piezas_coche = ["Motor","Suspensión","Frenos","Transmisión","Escape","Refrigeración","Electrónica","Iluminación","Carrocería","Neumáticos"]
colores = ['Rojo', 'Azul', 'Verde', 'Amarillo', 'Negro', 'Blanco', 'Gris', 'Naranja','Marrón', 'Blanco', 'Plateado' ]
```

```python
#Creamos una función para llamar a las piezas como una categoría con 3 dígitos seguidos
def generar_nombre(categoria):
    return f"{categoria}-{random.randint(100, 999)}"
```

```python
# Crear tabla "pieza"
query = '''
CREATE TABLE IF NOT EXISTS pieza (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT,
    codigo_pieza TEXT,
    color TEXT,
    precio REAL,
    categoria TEXT
)
'''
cursor_prov.execute(query)
```

```python
faker = Faker('es_ES')
```

```python
#Creamos una lista donde meteremos los datos de nuestra tabla "pieza"
piezas_info_list = []
```

```python
#Hacemos la lista piezas y añadimos nuestros valores
#Nuestra categoría tiene que ser igual que nuestro nombre de pieza : (categoría)-XXX


for _ in range(30):
    categoria = random.choice(categorias_piezas_coche)
    nombre = generar_nombre(categoria)
    codigo_pieza = f"{categoria[:2].upper()}-{random.randint(10000, 99999)}"
    color = random.choice(colores)
    precio = round(random.uniform(20, 1500), 2)

    piezas_info_list.append((nombre, codigo_pieza,color,precio, categoria ))

    cursor_prov.execute('''
    INSERT INTO pieza (nombre, codigo_pieza, color, precio, categoria)
    VALUES (?, ?, ?, ?, ?)
    ''', (nombre, codigo_pieza, color, precio, categoria))

print(f"codigo_pieza: {codigo_pieza}, nombre: {nombre}")
```
