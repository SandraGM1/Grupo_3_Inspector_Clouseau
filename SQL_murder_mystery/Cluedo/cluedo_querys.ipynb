```python
import pandas as pd
import sqlite3
```

```python
cnx = sqlite3.connect("sql-murder-mystery.db")
```

```python
'''Empezamos seleccionando  nuestra tablas de nuetsra base de datos '''

query = ''' 
SELECT *
FROM sqlite_master
where type = "table"
'''

pd.read_sql(query, cnx)
```

```python
'''Empezamos buscando por la fecha y ciudad en el crime scene report'''

query = ''' 
SELECT *
FROM crime_scene_report
where date = "20180115" and type = "murder" and city = "SQL City"
'''
pd.read_sql(query, cnx)
```

```python
'''Seguimos, ahora buscando por la dirección que nos daban de nuestros testigos, Northwestern Dr '''
query = ''' 
SELECT *
FROM person
WHERE address_street_name in ("Northwestern Dr")
ORDER BY address_number DESC   
'''
pd.read_sql(query, cnx)
'''Morty Shapiro es nuestro primer testigo

'''
```
